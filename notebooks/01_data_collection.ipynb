```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
stocks = ['AAPL', 'MSFT', 'NVDA', 'AMZN', 'GOOGL']
```

```python
data = yf.download(
    stocks,
    start = '2019-01-01',
    end = '2025-01-01'
)['Close']
```

```python
data.head() #stock prices
```

```python
print(data.shape)

data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize = (12, 6))

for stock in stocks:
    plt.plot(data[stock], label = stock)

plt.title('Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()

plt.savefig('../images/stock_prices.png',
            bbox_inches='tight')
plt.show()
```

```python
#saving data
data.to_csv('../data/stock_prices.csv')
```
